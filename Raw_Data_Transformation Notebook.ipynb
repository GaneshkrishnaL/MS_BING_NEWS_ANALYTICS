```python
df = spark.read.option("multiline", "true").json("Files/latest-bing-news.json")
# df now is a Spark DataFrame containing JSON data from "Files/latest-bing-news.json".
display(df)
```

```python
df=df.select('Value').limit(200)
```

```python
display(df)
```

```python
from pyspark.sql.functions import explode
dfexploded= df.select(explode(df['Value']).alias('json_object'))
```

```python
display(dfexploded)
```

```python
#each row is converted to json doc using .toJSON() and also converted to list using .collect()

listofnews=dfexploded.toJSON().collect()
```

```python
#now convert these string json docs into dictionary to easily retrive data using json.loads()
```

```python
# name , category, description , provider name, image url, urlor article, date published

print(jsondict["json_object"]["name"])
print(jsondict["json_object"]["category"])
print(jsondict["json_object"]["description"])
print(jsondict["json_object"]["provider"][0]["name"])
print(jsondict["json_object"]["image"]["thumbnail"]["contentUrl"])
print(jsondict["json_object"]["url"])
print(jsondict["json_object"]["datePublished"])
```

```python
title=[]
category=[]
description=[]
provider=[]
imageurl=[]
articleurl=[]
dateposted=[]

for i in listofnews:


    try:
        newsdict=json.loads(i)

        if newsdict["json_object"].get("category") and newsdict["json_object"].get("image",{}).get("thumbnail",{}).get("contentUrl") :

            title.append(newsdict["json_object"]["name"])
            category.append(newsdict["json_object"]["category"])
            description.append(newsdict["json_object"]["description"])
            provider.append(newsdict["json_object"]["provider"][0]["name"])
            imageurl.append(newsdict["json_object"]["image"]["thumbnail"]["contentUrl"])
            articleurl.append(newsdict["json_object"]["url"])
            dateposted.append(newsdict["json_object"]["datePublished"])


    except Exception as e:
        print('Error in : {}'.format(e))
```

```python
title
```

```python
#now coverting all these lists into the columns in out data frame, and define schema - giving column name, and data type

from pyspark.sql.types import StringType,StructField,StructType

combinedlists=list(zip(title,category,description,provider,imageurl,articleurl,dateposted))

schema= StructType([
    StructField("title",StringType(),True),
    StructField("category",StringType(),True),
    StructField("description",StringType(),True),
    StructField("provider",StringType(),True),
    StructField("imageurl",StringType(),True),
    StructField("articleurl",StringType(),True),
    StructField("dateposted",StringType(),True)
]
)

properdata = spark.createDataFrame(combinedlists,schema=schema)
```

```python
display(properdata)
```

```python
from pyspark.sql.functions import to_date , date_format

clean_df=properdata.withColumn("dateposted",date_format(to_date("dateposted"),"dd-MMM-yyyy"))
```

```python
display(clean_df)
```

## SETTING INCREMENTAL LOAD 

```python
from pyspark.sql.utils import AnalysisException

try:
    name="bing_news_lakehouse.tbl_latestnews"
    clean_df.write.format("delta").saveAsTable(name)
except AnalysisException:
    print('Table Already Exists')
    clean_df.createOrReplaceTempView("viewOfCleandf")

    spark.sql(f""" MERGE INTO {name} target_table
                USING viewOfCleandf source_table

                ON target_table.articleurl=source_table.articleurl

                WHEN MATCHED AND 

                source_table.title <> target_table.title OR
                source_table.category <> target_table.category OR
                source_table.description <> target_table.description OR
                source_table.provider <> target_table.provider OR
                source_table.imageurl <> target_table.imageurl OR
                source_table.dateposted <> target_table.dateposted

                THEN UPDATE SET * 

                WHEN NOT MATCHED THEN INSERT * 
             """)

```
